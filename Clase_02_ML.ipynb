# **Repaso de Pandas, Matplotlib, Seaborn y Plotly**


*   *Pandas*: Trabajar Dataframes
*   *Matplotlib* y Seaborn: Crear gráficos
*   *Plotly*: Gráficos interactivos


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub #Acceso a kaggle
import os        #Para las rutas de archivos

#Configuro el estilo de seaborn con fondo blanco y cuadricula para leer mejor los datos
sns.set_style("whitegrid")

#Ajusto el tamaño estandar de los graficos (8 pulgadas de ancho 5 de alto)

plt.rcParams['figure.figsize'] = (8,5)
```

# Cargo el dataset

```python
ruta_dataset = kagglehub.dataset_download("rohitsahoo/sales-forecasting")
ruta_csv = os.path.join(ruta_dataset, "train.csv")

df = pd.read_csv(ruta_csv)
print("Se descargo correctamente el dataset de Kaggle")
```

# Exploracion y limpieza basica con **Pandas**

```python
# 1. Informacion general: tipos de datos y cantidad de valores nulos
print("Tipos de datos y cantidad de valores nulos: \n")
df.info()
print()
print("-"*60)

# 2. Conteo de valores nulos para ver cuantos nos falta
print("Conteo de valores nulos: \n")
print(df.isnull().sum())
print()
print("-"*60)

# 3. Resumen estadístico: vemos medias, máximos, mínimos y más
# (solo sobre columnas numéricas)
print("Resumen estadístico:\n")
display(df.describe())
print()
print("-"*60)

# 4. Filtrado: nos quedamos solo con las ventas mayores a 1000
print("Ventas mayores a 1000:\n")
display(df[df['Sales'] > 1000].head())
print()
print("-"*60)

# 5. Verificamos cuál es el valor más alto en la columna Sales
print("La venta más alta registrada es de:", df['Sales'].max())
```

# Resumen de comandos:

*   **df.info():** Es como el "DNI" del dataset. Te dice cuántas filas tenés, qué columnas hay y, lo más importante, si Python las reconoce como números, texto (objetos) o fechas.

*   **df.isnull().sum()**: Fundamental. Si una columna tiene muchos valores nulos, el modelo de Machine Learning puede fallar o dar resultados erróneos. Acá identificamos esos "huecos".

*   **df.describe():** Te da un pantallazo estadístico. Si ves que el valor mínimo de Sales es negativo, por ejemplo, ya sabés que tenés datos que no tienen sentido y vas a tener que revisarlos.

*   **df[df['Sales'] > 1000]:** Esta es la forma en la que le pedimos a pandas que filtre. Estamos haciendo una pregunta: "¿Cuáles son las filas donde la columna Sales es mayor a 1000?". Es una herramienta muy poderosa para encontrar casos particulares o valores extremos (outliers).



## Matplotlib - Gráficos básicos


```python
# 1. Histograma: Para ver cómo se distribuyen nuestras ventas
plt.hist(df['Sales'],
         bins=30,
         color='skyblue')
plt.title('Distribución de las ventas')
plt.xlabel('Ventas (USD)')
plt.ylabel('Frecuencia')
# usar plt.xlim(min, max) para hacer "zoom"
plt.show()
```

```python
# 2. Gráfico de barras: Para comparar cantidades entre categorías
cat_counts = df['Category'].value_counts()
plt.bar(cat_counts.index,
        cat_counts.values,
        color='coral')
plt.title('Ventas por categoría')
plt.xlabel('Categoría')
plt.ylabel('Cantidad')
plt.show()
```

## Seaborn - Visualizaciones estadísticas

Ahora vamos a dar un salto de calidad con **Seaborn**. Esta librería está construida sobre **Matplotlib**, pero está diseñada específicamente para estadísticas y gráficos más complejos (y un poco mas modernos esteticamente) con menos líneas de código.

Vamos a realizar dos análisis muy importantes:

```python
# 1. Boxplot (Diagrama de caja):
# Para ver la dispersión de las ventas por categoría
sns.boxplot(x='Category', y='Sales', data=df)
plt.title('Boxplot de ventas por categoría')
plt.show()
```

```python
# 2. Heatmap (Mapa de calor):
# Para entender la relación entre las variables
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
```

# ¿Qué estamos mirando en estos gráficos?


*   El Boxplot: Es uno de los gráficos más importantes en Machine Learning. Nos muestra la mediana (la línea dentro de la caja), los cuartiles (donde se concentra el grueso de los datos) y, sobre todo, esos puntitos que sobran arriba: los valores atípicos o outliers. Si antes te preguntabas por qué el gráfico de ventas se "estiraba" tanto, acá vas a ver a los culpables.

*   El Heatmap de correlación: Este gráfico nos dice qué tanto se "parecen" o se influyen dos variables entre sí.

      *   Un valor cercano a 1 (color cálido/rojo) significa que cuando una variable sube, la otra también.
      *   Un valor cercano a -1 (color frío/azul) significa que cuando una sube, la otra baja.
      *   Un valor cercano a 0 significa que no hay una relación clara.

Nota: Usamos numeric_only=True porque no tiene sentido calcular correlaciones entre palabras (como nombres de ciudades), solo lo hacemos entre números.

## Seaborn: Countplot (Conteo de registros)

A veces no queremos sumar valores, solo queremos saber cuántas veces aparece cada elemento. Para eso usamos `countplot()`. Es una forma rapidísima de ver qué subcategorías tienen más movimiento.

Fijate que usamos `order` para que las barras aparezcan ordenadas de mayor a menor; esto hace que el gráfico sea mucho más fácil de leer que si estuvieran desordenadas.

```python
# Contamos cuántos registros hay por cada subcategoría
sns.countplot(
    x='Sub-Category',
    data=df,
    order=df['Sub-Category'].value_counts().index)

# Rotamos las etiquetas del eje X 45 grados para que no se encimen los nombres
plt.xticks(rotation=45)

plt.title('Cantidad de registros por Subcategoría')
plt.show()
```

## Visualización 1: distribución de una variable numérica

Una buena forma de comenzar a visualizar los datos es observar cómo se distribuye una variable numérica.

En este caso, vamos a trabajar con el largo del pétalo. Un histograma nos permitirá ver en qué rangos se concentran más observaciones.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="PetalLengthCm", bins=20, kde=True)
plt.title("Distribución del largo del pétalo")
plt.xlabel("Largo del pétalo")
plt.ylabel("Frecuencia")
plt.show()
```

Este gráfico permite detectar la forma general de la distribución.

Además, aparecen varios picos. Eso puede sugerir que las distintas especies no se comportan igual respecto de esta variable.

## Visualización 2: relación entre dos variables

Ahora vamos a representar la relación entre dos variables numéricas mediante un gráfico de dispersión.

Usaremos el largo y el ancho del pétalo, y además diferenciaremos los puntos según la especie para ver si aparecen agrupamientos.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="PetalLengthCm",
    y="PetalWidthCm",
    hue="Species",
    s=70
)

plt.title("Relación entre largo y ancho del pétalo")
plt.xlabel("Largo del pétalo")
plt.ylabel("Ancho del pétalo")
plt.show()
```

En este gráfico se puede observar si existe relación entre ambas variables y si las especies forman grupos distinguibles.

Cuando los puntos de distintas categorías se separan con cierta claridad, eso sugiere que esas variables pueden ser útiles para diferenciar clases.

## Visualización complementaria: comparación por especie

Además del gráfico de dispersión, también es útil comparar una variable numérica entre categorías.

Para eso podemos usar un boxplot, que resume la distribución de cada grupo y permite comparar medianas, dispersión y posibles valores extremos.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Species", y="PetalLengthCm")

plt.title("Largo del pétalo según especie")
plt.xlabel("Especie")
plt.ylabel("Largo del pétalo")
plt.show()
```

Este tipo de gráfico ayuda a comparar rápidamente cómo varía una medida numérica entre grupos.

En particular, permite ver si una especie tiende a tener valores más altos o más bajos que las otras, y si presenta mayor o menor dispersión.
